```python
nb = int(np.log2(cutoff))
n = 1 + nb
basis = [0]*n + [1] + [0]*nb + [0]*n
fermion_qubits = [n]

target_fn = sum(basis[i] for i in fermion_qubits)


def fermion_number_from_bitstring(bitstr: str, fermion_qubits):
    # Reverse so index 0 corresponds to q0 (like your basis list)
    bits = bitstr[::-1]
    return sum(int(bits[i]) for i in fermion_qubits)

def filter_counts_by_fermion_number(counts: dict[str, int],
                                    basis: list[int],
                                    fermion_qubits: list[int]) -> dict[str, int]:
    target_fn = sum(basis[i] for i in fermion_qubits)

    filtered = {
        bitstr: c
        for bitstr, c in counts.items()
        if fermion_number_from_bitstring(bitstr, fermion_qubits) == target_fn
    }
    return filtered

```
