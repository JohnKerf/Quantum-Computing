```python
import os, json
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ns = [2,3,4]
N=4
cts = ["QFT_full", "QFT_trunc", "Trunc", "Fock", "Position"]
cutoffs = [2,4,8,16,32,64]
potential= "linear"

all_data = {}

for basis in cts:
    data = []
    for cutoff in cutoffs:

        path = os.path.join(r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\Wess-Zumino\Qiskit\SKQD\GateScaling\BasisScaling\ibm_torino\partials2\lie", f"linear_N{N}_L{cutoff}_{basis}.json")
       
        try:
            with open(path, encoding="utf-8") as f:
                d = json.load(f)
            data.append(d['cost'])
        except:
            data.append(None)


    all_data[basis] = data
```

```python
potential = "linear"
cutoffs = [2**n for n in range(1,7)]

num_qubits = [N*int(1+np.log2(c)) for c in cutoffs]
#scaling_4 = [4**(N*int(1+np.log2(c))) for c in cutoffs]
#scaling_2 = [2**(N*int(1+np.log2(c))) for c in cutoffs]
scaling_3 = [(N*int(1+np.log2(c)))**2 for c in cutoffs]

metric_key = "num_2q_ops" #"num_2q_ops"
#all_data['$4^{N_{qubits}}$'] = scaling_4
#all_data['$2^{N_{qubits}}$'] = scaling_2
all_data['$N^{2}$'] = scaling_3

plt.figure(figsize=(8, 5))
for label, y in all_data.items():
    plt.plot(num_qubits, y, marker="o", linewidth=2, label=label)

plt.yscale("log")
#plt.xscale("log")
plt.xlabel("$N_{qubits}$")
plt.ylabel("2Q Gates")
plt.title(f"{potential}")
plt.grid(True, which="both", linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
