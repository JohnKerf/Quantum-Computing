```python
import pandas as pd
import os
import json
```

```python
folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\PennyLane\ADAPT-VQE\Files"
folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
```

```python
cut_off_list = [2, 4, 8, 16, 32, 64]

potentials = []
cutoffs = []
basis_states = []
op_lists = []
min_e_exact = []
min_e = []

for f in folders:
    for cutoff in cut_off_list:
        data_path = folder_path + "\\{}\\{}_{}.json".format(f,f, cutoff)
                
        with open(data_path, 'r') as file:
            data = json.load(file)

        potentials.append(data['potential'])
        cutoffs.append(data['cutoff'])
        basis_states.append(data['basis_state'])
        op_lists.append(data['op_list'])
        min_e_exact.append(data['min_e_exact'])
        min_e.append(data['min_e'])
```

```python
df = {"potentials":potentials,
    "cutoffs":cutoffs,
    "basis_states":basis_states,
    "op_lists":op_lists,
    "min_e_exact":min_e_exact,
    "min_e":min_e}
```

```python
df = pd.DataFrame(df)
df
```
