```python
import numpy as np
import pandas as pd
import json
import os
from tabulate import tabulate

import matplotlib.pyplot as plt
```

```python
folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\Qiskit-Vs-PL_pauli\PL\Files"
folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\Qiskit-Vs-PL_pauli\PL\FilesADAPT-OpParams"
#folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\Qiskit-Vs-PL_pauli\PL\FilesADAPT"
#folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\Qiskit-Vs-PL_pauli\Qiskit\Files"
dataframes = {}
folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
folders
```

```python
cutoff_list = [2,4,8,16]
colours = ['red', 'blue', 'green', 'orange']
for potential in folders:
    all_data = []
    eigs = []
    for cutoff in cutoff_list:

        d_path = os.path.join(folder_path, potential, f'{potential}_{cutoff}.json')

        with open(d_path, 'r') as json_file:
            data = json.load(json_file)

        converged_indices = [i for i, success in enumerate(data['success']) if success]
        results = [data['results'][i] for i in converged_indices]

        all_data.append(results)
        eigs.append(data['exact_eigenvalues'][0])

    plt.figure(figsize=(15,8))
    plt.boxplot(all_data, vert=False, tick_labels=cutoff_list)

    for eig, c, colour in zip(eigs, cutoff_list, colours):
        plt.axvline(x=eig, color=colour, linestyle="--", linewidth=1.0, alpha=0.7, label=f"$\\Lambda$={c}")

    plt.title(f"{potential}")
    plt.xscale('symlog', linthresh=1e-5)
    plt.xlabel("Energy")
    plt.ylabel("$\\Lambda$")
    plt.legend()
    plt.grid(axis="x", linestyle="--", alpha=0.7)

    save_path = os.path.join(folder_path, potential, "box_plot.png")
    plt.savefig(save_path)
```
