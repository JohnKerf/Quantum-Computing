```python
import numpy as np
import pandas as pd
from susy_qm import calculate_Hamiltonian
```

```python
num_cut_offs = 12
cutoffs = [2**i for i in range(1, num_cut_offs + 1)]
potentials = ['QHO', 'AHO', 'DW']
data = []
all_data = []

for potential in potentials:

    print(f"Running for {potential} potential")
    e0 = []
    e1 = []
    e2 = []
    ratios = []

    for cutoff in cutoffs:

        H = calculate_Hamiltonian(cutoff, potential)
        eigenvalues = np.sort(np.linalg.eig(H)[0])[:3].real

        e0.append(eigenvalues[0])
        e1.append(eigenvalues[1])
        e2.append(eigenvalues[2])

        r = (eigenvalues[2] - eigenvalues[1]) / (eigenvalues[2] - eigenvalues[0])
        ratios.append(r)

    data = {"cutoff": cutoffs,
        "e0": e0,
        "e1": e1,
        "e2": e2,
        "ratios": ratios}

    all_data.append(data) 
```
