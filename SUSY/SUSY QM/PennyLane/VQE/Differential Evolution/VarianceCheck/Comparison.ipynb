```python
import numpy as np
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
```

```python
potential = "QHO"
folder_path = r"C:\Users\Johnk\Documents\PhD\Quantum Computing Code\Quantum-Computing\SUSY\SUSY QM\PennyLane\VQE\Differential Evolution\VarianceCheck\Files\{}".format(potential)
cutoffs = [16]
folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]
```

```python
folders = sorted([int(f) for f in folders])
```

```python
mean_variances = []
std_variances = []
for f in folders:

    for cutoff in cutoffs:

        data_path = os.path.join(folder_path, str(f), "{}_{}.json".format(potential, cutoff))
        
        with open(data_path, 'r') as file:
            data = json.load(file)

        variances = [var for vars in data['variances'] for var in vars]
        mean_v = np.mean(variances)
        std_v = np.std(variances)
        
        mean_variances.append(mean_v)
        std_variances.append(std_v)
```

## AHO

```python
plt.figure(figsize=(24, 10))
plt.errorbar(folders, mean_variances, yerr=std_variances, fmt='o-', capsize=5)
plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Shots')
plt.ylabel('Mean Variance')
plt.title('Mean Variance vs. Shots')
plt.grid(True)
plt.show()
```
