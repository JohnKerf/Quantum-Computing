```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = np.linspace(-2, 2, 400)
m = 1
W = 0.5 * m * q**2

plt.plot(q, W, label=r"$W(q) = \frac{1}{2} m q^2 + \frac{1}{4} g q^4$")
plt.xlabel(r"$q$")
plt.ylabel(r"$W(q)$")
plt.title("Anharmonic Superpotential (Supersymmetric AHO)")
#plt.legend()
#plt.grid(True)
plt.show()
```

```python
q = np.linspace(-2, 2, 400)
m, g = 1, 1
W = 0.5 * m * q**2 + 0.25 * g * q**4

plt.plot(q, W, label=r"$W(q) = \frac{1}{2} m q^2 + \frac{1}{4} g q^4$")
plt.xlabel(r"$q$")
plt.ylabel(r"$W(q)$")
plt.title("Anharmonic Superpotential (Supersymmetric AHO)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
q = np.linspace(-100, 100, 400)
m, g, mu = 1, 1, 1
W = 0.5 * m * q**2 + g*(1/3 * q**3 + mu**2 * q)

plt.plot(q, W, label=r"$W(q) = \frac{1}{2} m q^2 + \frac{1}{4} g q^4$")
plt.xlabel(r"$q$")
plt.ylabel(r"$W(q)$")
plt.title("Anharmonic Superpotential (Supersymmetric AHO)")
plt.legend()
plt.grid(True)
plt.show()
```
