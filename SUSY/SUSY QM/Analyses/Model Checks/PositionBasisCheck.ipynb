```python
import susy_qm as sqm
from qiskit.quantum_info import SparsePauliOp
import numpy as np
```

```python
cutoff = 2**7
potential="QHO"

H_pauli, num_qubits, H_dense = sqm.build_sqm_hamiltonian_position_basis_pauli(
    cutoff=cutoff,
    potential=potential,   # or "AHO", "DW"
    m=1.0, g=1.0, u=1.0,
    x_max=8.0,
    atol_from_operator=1e-12,
)

pauli_labels = H_pauli.paulis.to_labels()
pauli_coeffs = H_pauli.coeffs.tolist()
print(num_qubits, len(pauli_labels))

me = np.linalg.eigvalsh(H_pauli.to_matrix()).min().real
me
```

```python
H = sqm.calculate_Hamiltonian(cutoff, potential)

H_pauli = SparsePauliOp.from_operator(H)
pauli_labels = H_pauli.paulis.to_labels()
pauli_coeffs = H_pauli.coeffs.tolist()
print(num_qubits, len(pauli_labels))

me = np.linalg.eigvalsh(H_pauli.to_matrix()).min().real
me
```
